## "while" loop

```python
import random
```

```python
# Nim game 
# game rule 1: The player who takes the last one stick is winner
# game rule 2: Each player takes turns picking up 1, 2 or 3 sticks and cannot pass.
# please understand the following code to get used to "while"
total_sticks = 10

# only 2 players A, B
turn = 1
while total_sticks > 3:
    num_stick_taken = random.randint(1, 3)
    if turn:
        print ('A takes', num_stick_taken)
    else:
        print ('B takes', num_stick_taken)
    total_sticks -= num_stick_taken
    turn = 1 - turn
    
if turn:
    print ('A win')
else:
    print ('B win')
```
